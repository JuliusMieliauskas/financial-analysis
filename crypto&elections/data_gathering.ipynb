```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data is taken from the World's leading crypto exchange

```python
btc_df = pd.read_csv('data/BTCUSD.csv')
eth_df = pd.read_csv('data/ETHUSD.csv')
```

```python
btc_df['date'] = pd.to_datetime(btc_df['date'])
btc_df.set_index('date', inplace=True)
btc_df.sort_index()

eth_df['date'] = pd.to_datetime(eth_df['date'])
eth_df.set_index('date', inplace=True)
eth_df.sort_index()
```

```python
aligned_df = pd.merge(btc_df, eth_df, left_index=True, right_index=True, how='outer')
```

```python
aligned_df = aligned_df.loc[:, ['close_x', 'close_y', 'Volume USD_x', 'Volume USD_y']]
aligned_df.columns = ['BTC', 'ETH', 'BTC Volume in USD', 'ETH Volume in USD']
```

```python
aligned_df.to_csv('data/merged.csv')
```
